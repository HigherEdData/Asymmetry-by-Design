### Figure A2: Total undergraduate enrollment at for-profits by highest degree offered at college

```stata
ssc install gr0070
```

```stata
copy "https://github.com/HigherEdData/asymmetry/raw/master/data/d_unitidasymmetry.dta" ///
    data/d_unitidasymmetry.dta, replace
```

```stata
quietly {
set scheme plotplain
est clear

quietly use data/d_unitidasymmetry, clear
drop if iclevel==.
collapse (rawsum) all_under ///
, by (year iclevel)

reshape wide all_under, i(year) j(iclevel)
    
label var all_under1 "At least BA"
label var all_under2 "AA"
label var all_under3 "certificate"

drop if year>2015 | year<1990

graph tw (connect all_under* year, sort msym(D D D O O O O O) ///
	mc(gs10 black black) ///
	mfc(white gs12 gs8) ///
	lp(dot l dot) lc(black gs10 black) ///
	lw(medthick medthick medthick)), ///
	legend(size(small) pos(3) order(1 2 3) ///
	subtitle("{bf:Highest offering}", size(small))) ///
	ytitle("Total undergrad enrollment (thousands)") ylabel(,format(%10.0fc)) ///
	name(highestdegreetime, replace) xlab(1990(5)2015)
    
graph export figures/fa2_asymmetryenrollmenthighestdegree.pdf, replace
}
```
